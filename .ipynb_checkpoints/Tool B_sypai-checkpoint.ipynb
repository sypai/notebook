```python
def find_segments(inList):
    startIdx = 0
    endIdx = 0
    idx = 0
    outSeg = []
    checkBit = inList[startIdx]
    
    for item in inList:
        if item != checkBit:
            idx0 = startIdx
            idx1 = idx - 1
            startIdx = idx
            outSeg.append([idx0, idx1, checkBit])
            checkBit = item
        idx += 1
    
    return outSeg
```

**audSeg** & **subSeg** are different blocks in the audio and subtitles.

```python
audstring, substring = map(
            lambda s: 2 * np.array(s).astype(float) - 1, [audstring, substring])
```

```python
total_bits = math.log(len(audstring) + len(substring), 2)
total_length = int(2 ** math.ceil(total_bits))
extra_zeros = total_length - len(audstring) - len(substring)
audft = np.fft.fft(np.append(np.zeros(extra_zeros + len(substring)), audstring))
subft = np.fft.fft(np.flip(np.append(substring, np.zeros(len(audstring) + extra_zeros)), 0))
convolve = np.real(np.fft.ifft(audft * subft))
best_idx = np.argmax(convolve)
score = len(convolve) - 1 - best_idx - len(audstring)
print(score)
```
